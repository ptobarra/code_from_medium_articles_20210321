```python
# let's run everything together

import requests
import re
import numpy
import pandas
from pandas import DataFrame
from bs4 import BeautifulSoup
from tqdm import tqdm_notebook

def extract_books_from_result(soup):
  returner = {'books': [], 'authors': []} # definimos diccionario donde guardar libros y autores

  for book in soup.find_all('div', attrs={'class': 'cmp-text text'}):  
    text=book.text  
    pattern = re.compile(r'(?<=. ).+(?=\n)') # nos quedamos con loq vaya despues del punto y antes del salto linea
                                           # the ‘.+’ expression in the middle basically means ‘bring me anything’
    appender = re.findall(pattern,text)[0].split('by')

    # Including a condition just to avoid anything that’s not a book
    # Each book should consist in a list with two elements: 
    # [0] = title and [1] = author

    if len(appender) > 1:
      returner['books'].append(appender[0])
      returner['authors'].append(appender[1])

  returner_df = DataFrame(returner, columns=['books','authors'])

  authors=returner_df['authors'].tolist()

  returner_df.drop(['authors'], axis=1)

  for i in range (0, len(authors)):
    authors[i]=authors[i].strip() # strip() method removes whitespace at the \
                                  # beginning and end (both sides) of a string.  
  
  returner_df['authors']=authors
    
  return returner_df

url = 'https://www.penguin.co.uk/articles/2018/100-must-read-classic-books/'

r = requests.get(url)

soup = BeautifulSoup(r.text,'html.parser')

results = extract_books_from_result(soup)

print(results)
```
